```python
df = pd.read_csv('7.1 car-sales.csv')
df
```

```python
#removing the extra zeros after the point from the price column
df['Price'] = df['Price'].str[:-2]
df
```

```python
#removing the dollar sign from the prices column
df['Price'] = df['Price'].str.replace('[\$\,\.]', '').astype(int)
df
```

```python
#inputting a new column called sale date using date range
df['sale date'] = pd.date_range('10/10/2020', periods=len(df))
df
```

```python
#adding a new column called total sales, using `cumsum` function
df['total sales'] = df['Price'].cumsum()
df
```

```python
#plotting total sales against sale date
df.plot(x='sale date', y='total sales');
```

## The plotting method used previously is the pyplot method
## now we are going to look at the `OO method` which means object orientation method

```python
heart_disease.head()
```

```python
#selecting the data of age from 50 and above from the dataset
over_50 = heart_disease[heart_disease['age']>50]
len(over_50)
```

```python
over_50.head()
```

```python
# OO METHOD mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',#this is the color, using the target values
             ax = ax);
             
#ax.set_xlim([45, 100]); #this increases the space around the axes(which is the area of the plot)
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
#customize the plot
ax.set(title='heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol')

#add legend
ax.legend(*scatter.legend_elements(), title='target')

#adding a horizontal line accross to show the average cholesterol level(mean line)
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
#subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                               nrows=2,
                               ncols=1,
                               sharex=True)#this shares the xlabel to avoid repetition

#add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

#customize ax0
ax0.set(title='heart disease and cholesterol levels',
        ylabel='cholesterol')
#add legend to ax0
ax0.legend(*scatter.legend_elements(), title='target');

#adding a horizontal line accross to show the average cholesterol level(mean line)
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

#add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
#customize ax1
ax1.set(title='heart disease and max heart rate levels',
        xlabel='Age',
        ylabel='thalach (max heart rate)')
#add legend to ax1
ax1.legend(*scatter.legend_elements(), title='target');

#adding a horizontal line accross to show the average max heart rate level(mean line)
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--')

#add title to the Figure
fig.suptitle('HEART DISEASE ANALYSIS', fontsize=16, fontweight='bold');
```

```python
#set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter') #this changes the colour scheme
#customize the plot
ax.set(title='heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol')

#add legend
ax.legend(*scatter.legend_elements(), title='target')

#adding a horizontal line accross to show the average cholesterol level(mean line)
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
#customizing the x and y limitations


#subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                               nrows=2,
                               ncols=1,
                               sharex=True)#this shares the xlabel to avoid repetition

#add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                    cmap='winter')

#customize ax0
ax0.set(title='heart disease and cholesterol levels',
        ylabel='cholesterol')
#change the x axis limit
ax0.set_xlim([50, 80])
#add legend to ax0
ax0.legend(*scatter.legend_elements(), title='target');

#adding a horizontal line accross to show the average cholesterol level(mean line)
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

#add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='summer')
#customize ax1
ax1.set(title='heart disease and max heart rate levels',
        xlabel='Age',
        ylabel='thalach (max heart rate)')
#change ax1 y axis limit
ax1.set_ylim([60, 200])
#add legend to ax1
ax1.legend(*scatter.legend_elements(), title='target');

#adding a horizontal line accross to show the average max heart rate level(mean line)
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--')

#add title to the Figure
fig.suptitle('HEART DISEASE ANALYSIS', fontsize=16, fontweight='bold');
```
